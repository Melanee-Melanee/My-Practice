# count function in string for one keyword


```python
#count function in string for one keyword

def count_keyword(text, keyword):
 # Convert both text and keyword to lower case for case-insensitive comparison
 text = text.lower()
 keyword = keyword.lower()
 
 # Initialize a counter
 count = 0
 # Split the text into words
 words = text.split()
 
 # Iterate through each word in the text
 for word in words:

 # Check if the current word matches the keyword
     if word == keyword:
    
        count += 1
 
 return count

# Example usage
text = "Python is great. I love Python programming. Python is versatile."
keyword = "python"
result = count_keyword(text, keyword)

print(keyword,":" ,result)
#print(f"The keyword '{keyword}' appears {result} times in the text.")

```

# count function in string for some keywords


```python
#count function in string for some keywords



def count_keywords(text, keywords):
     # Convert the text to lower case for case-insensitive comparison
    text = text.lower()

     # Initialize a dictionary to store the counts
    keyword_counts = {}
 
     # Count occurrences of each keyword using the count() method
    for keyword in keywords:
        
         # Convert the keyword to lower case
        keyword_lower = keyword.lower()
             # Use the count() method to find occurrences of the keyword in the text
        count = text.count(keyword_lower)
             # Store the count in the dictionary
        keyword_counts[keyword] = count

    return keyword_counts

# Example usage
text = "Python is great. I love Python programming. Python is versatile. Python is also used for data science."
keywords = ["python", "data", "great", "science", "java"]

result = count_keywords(text, keywords)

# Print the results
for keyword, count in result.items():
    print(keyword,":" ,count)
    #print(f"The keyword '{keyword}' appears {count} times in the text.")
```

# count function in string for some keywords in one pdf


```python
#count function in string for some keywords in one pdf

import os
from pdfminer.high_level import extract_text

def count_keywords_occurrences(file_path, keywords):
    try:
        # Extract text from the PDF file
        text = extract_text(file_path)

        # Convert the extracted text and keywords to lowercase for case-insensitive search
        text_lower = text.lower()

        # Count occurrences for each keyword
        occurrences = {keyword: text_lower.count(keyword.lower()) for keyword in keywords}

        # Print the results
        print(f"\nOccurrences in {file_path}:")
        for keyword, count in occurrences.items():
            print(keyword,':', count)
            #print(f"'{keyword}': {count} times")

    except Exception as e:
        print(f"Error: {e}")

# Example usage:
pdf_file_path = "/home/melanee/Downloads/Rahmani_CV_16.pdf"
search_keywords = ["python", "programming", "intelligence"]

count_keywords_occurrences(pdf_file_path, search_keywords)
```

# count function in string for some keywords in multiple pdfs


```python
#count function in string for some keywords in multiple pdfs

import os
import PyPDF2

def count_keywords_in_pdf(pdf_path, keywords):
    
    keyword_counts = {keyword: 0 for keyword in keywords}

    with open(pdf_path, 'rb') as file:
        
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            
            text = page.extract_text()
            if text: # Check if text extraction was successful
                
                text = text.lower() # Convert to lower case for case-insensitive matching
                for keyword in keywords:
                    
                    keyword_lower = keyword.lower()
                    keyword_counts[keyword] += text.count(keyword_lower)

    return keyword_counts

def count_keywords_in_directory(directory, keywords):
    
    for filename in os.listdir(directory):
        
        if filename.endswith('.pdf'):
            
            pdf_path = os.path.join(directory, filename)
            counts = count_keywords_in_pdf(pdf_path, keywords)
 
            # Print the results for this specific PDF
            #print(f"\nKeyword counts for PDF: {pdf_path}")
            print(pdf_path)
        
            for keyword, count in counts.items():
                
                #print(f"The keyword '{keyword}' appears {count} times.")
                print(keyword, ":", count)

# Example usage
directory_path = '/home/melanee/Downloads/' # Change this to your PDF directory
keywords = ['python', 'data', 'science', 'AI', 'machine learning']

count_keywords_in_directory(directory_path, keywords)
```

# count function in string for some keywords in multiple pdfs and sort them


```python
#count function in string for some keywords in multiple pdfs and sort them

import os
import PyPDF2

def count_keywords_in_pdf(pdf_path, keywords):
    
    keyword_counts = {keyword: 0 for keyword in keywords}

    with open(pdf_path, 'rb') as file:
        
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            
            text = page.extract_text()
            if text: # Check if text extraction was successful
                
                text = text.lower() # Convert to lower case for case-insensitive matching
                for keyword in keywords:
                    
                    keyword_lower = keyword.lower()
                    keyword_counts[keyword] += text.count(keyword_lower)

    return keyword_counts

def count_keywords_in_directory(directory, keywords):
    
    pdf_keyword_counts = []
    
    for filename in os.listdir(directory):
        
        if filename.endswith('.pdf'):
                
            pdf_path = os.path.join(directory, filename)
            counts = count_keywords_in_pdf(pdf_path, keywords)
            total_count = sum(counts.values())
            pdf_keyword_counts.append((pdf_path, counts, total_count))
            
    pdf_keyword_counts.sort(key=lambda x: x[2], reverse=True)
    
        # Print sorted results
    for pdf_path, counts, total_count in pdf_keyword_counts:
           
        print(f" PDF: {pdf_path} (Total Keywords: {total_count})")
        for keyword, count in counts.items():
                
            #print(f"The keyword '{keyword}' appears {count} times.")
            print(keyword, ":", count)


# Example usage
directory_path = '/home/melanee/Downloads/' # Change this to your PDF directory
keywords = ['python', 'data', 'science', 'AI', 'machine learning']

count_keywords_in_directory(directory_path, keywords)
```
